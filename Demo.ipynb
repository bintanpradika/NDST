```python
import vaex
#Download the dataset (12gb to your local/ public cloud machine from https://vaex.s3.us-east-2.amazonaws.com/taxi/yellow_taxi_2015_f32s.hdf5)
#Or stream the dataset directly from S3 (cached locally):
#df = vaex.open('s3://vaex/taxi/yellow_taxi_2015_f32s.hdf5?anon=true')
df = vaex.open('yellow_taxi_2015_f32s.hdf5') #open .hdf5 if the file exist in your local machine.
```

```python
df.describe()
```

```python
df
```

```python
#pickup visualization
long_min,long_max = min(df.pickup_longitude.values), max(df.pickup_longitude.values)
lat_min, lat_max = min(df.pickup_latitude.values), max(df.pickup_latitude.values)
df.plot(df.pickup_longitude, df.pickup_latitude, f="log1p", limits=[[long_min, long_max], [lat_min, lat_max]], show=True)

long_min,long_max = -74.05, -73.75
lat_min, lat_max = 40.6, 40.9
df.plot(df.pickup_longitude, df.pickup_latitude, f="log1p", limits=[[long_min, long_max], [lat_min, lat_max]], show=True)
```

```python
df.drop('pickup_datetime', inplace=True, check=True)
df.drop('dropoff_datetime' , inplace=True, check=True)
```

```python
df_final = df
df_final
```
